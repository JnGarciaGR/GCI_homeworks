<h1> Explicación función: </h1>

group_by: Para este caso, se agrupa la columna Score con la columna Type

Obteniendo así, un objeto tipo "groupby" y una agrupación del estilo:

````python
anime_data_extracted.groupby('Type')['Score']
````

<b><h3>Antes</h3></b>

````python
df = pd.DataFrame({
    "Tipo": ["TV", "TV", "Movie", "Movie", "OVA"],
    "Score": [8.0, 6.0, 7.0, 5.0, 9.0]
})
````

<b><h3>Después</h3></b>

Al aplicar .mean() se calculan sus respectivos promedios de cada Type-Score:

````python
anime_data_extracted.groupby('Type')['Score'].mean()
````

Obteniendo así:

TV = 7

Movie = 6

OVA = 9

````python
# Para que vaya de forma descendente
anime_data_extracted.groupby('Type')['Score'].mean().sort_values(ascending=False)
````
